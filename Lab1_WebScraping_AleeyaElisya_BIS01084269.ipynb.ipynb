```python
# ==========================================
# Name: Aleeya Elisya
# Student ID: BIS01084269
# CISB5123 Text Analytics
# Lab Assignment 1 - Web Scraping
# ==========================================


from google_play_scraper import reviews, Sort
import pandas as pd

def scrape_reviews(app_id, total_reviews):
    """
    This function scrapes reviews from a Google Play app.
    
    Parameters:
    app_id (str): The app package name from Google Play
    total_reviews (int): Number of reviews to collect
    
    Returns:
    DataFrame: Contains reviewer name, review date, and review content
    """
    
   
    result, continuation_token = reviews(
        app_id,
        lang='en',           
        country='us',        
        sort=Sort.NEWEST,    
        count=total_reviews 
    )
    
    
    reviewer_names = []
    review_dates = []
    review_contents = []
    
    
    for review in result:
        reviewer_names.append(review['userName'])
        review_dates.append(review['at'])
        review_contents.append(review['content'])
    
  
    data = pd.DataFrame({
        "Reviewer Name": reviewer_names,
        "Review Date": review_dates,
        "Review Content": review_contents
    })
    
    return data



app_id = "https://play.google.com/store/apps/details?id=com.shopee.my"


data = scrape_reviews(app_id, 100)


data.head()


data.to_csv("GooglePlay_Reviews.csv", index=False)

print("CSV file saved successfully!")
```
